```python
import pandas as pd
#Reading plk file into dataframe
apl_mf_pckl = pd.read_pickle('apple_mf_df_renamed.plk')
apl_mf_df=pd.DataFrame(apl_mf_pckl)
apl_mf_df.head()
```

```python
apl_mf_df["combined_data"]=apl_mf_df['mf_title'].map(str) +" "+ apl_mf_df['mf_text'].map(str) 
```

```python
apl_mf_df = apl_mf_df.sort_values("timestamp")
# apl_mf_df = apl_mf_df.set_index("timestamp")
apl_mf_df.head()
```

```python
import re
def text_clean(data): 
    #cleans the data and converts number to POSNUM or NEGNUM
    p='POSNUM'
    n='NEGNUM'
    data = data.lower()
    data = re.sub(r"what's", "what is ", data)
    data = re.sub(r"\'s", " is ", data)
    data = re.sub(r"\'d"," would ",data)
    data = re.sub(r"\'ve", " have ", data)
    data = re.sub(r"\'ll", " will ", data)
    data = re.sub(r"\'re", " are ", data)
    data = re.sub(r"can't", "cannot ", data)
    data = re.sub(r"n't", " not ", data)
    data = re.sub(r"i'm", "i am ", data)
    data = re.sub(r'-\d*\.?\d',n,data)
    data = re.sub(r'\d*\.?\d',p,data)
    data = re.sub('[^A-Za-z0-9 ]+', '', data)
    data = data.strip()
    
    return data
```

```python
from nltk.corpus import stopwords
def stopwords_removal(tokens):
    #removes top words
    final_sentence=[]
    stopwords_list=stopwords.words('english')
    for w in tokens:
        if w not in stopwords_list:
            final_sentence.append(w)
    return final_sentence
```

```python
from nltk.tokenize import RegexpTokenizer
regtokenizer = RegexpTokenizer(r'\w+')
def tokenize(x): 
    #generates tokens
    return regtokenizer.tokenize(x)
```

```python
from nltk.stem import WordNetLemmatizer
def wlemmatizer(x):
    #lemmatizes words
    lemm=WordNetLemmatizer()
    return ' '.join(lemm.lemmatize(w) for w in x)
```

```python
#first clean the data
apl_mf_df["combined_data"]=apl_mf_df["combined_data"].map(text_clean)
#then get tokens
apl_mf_df["tokens"]=apl_mf_df["combined_data"].map(tokenize)
#then remove stopwords
apl_mf_df["tokens"]=apl_mf_df["tokens"].map(stopwords_removal)
#then lemmatize
apl_mf_df["lemmatized"]=apl_mf_df["tokens"].map(wlemmatizer)
```

```python
apl_mf_df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
def senti_analysis(data):
    #generates polarity score
    return(analyser.polarity_scores(data)["compound"])
```

```python
apl_mf_df["lemm_sents"] = apl_mf_df["lemmatized"].apply(senti_analysis) 
```

```python
apl_mf_df["fulltext_sents"] = apl_mf_df["combined_data"].apply(senti_analysis) 
```
